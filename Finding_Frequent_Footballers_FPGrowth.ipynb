```python
import pandas as pd
import numpy as np
```

```python
df=[['Messi','Suarez','Ibrahimovic','Neymar','Ineista','Pique'],
    ['Neymar','Messi','Xavi','Ineista','Suarez','Fabrigas'],
    ['Xavi','Messi','Ineista','Neymar','Ter stegan','Ineista'],
    ['Dani alves','Mascherano','Fabrigas','Ter stegan','Suarez','Pique'],
    ['Pique','Neymar','Messi','Pique','Fabrigas','Dani alves']]
```

```python
#Transaction_Encoder
from mlxtend.preprocessing import TransactionEncoder
t=TransactionEncoder()
t_d=t.fit_transform(df)
df=pd.DataFrame(t_d,columns=t.columns_)
```

```python
df
```

```python
#fpgrowth_algorithm
from mlxtend.frequent_patterns import fpgrowth
fpgrowth(df,min_support=0.3)
```
